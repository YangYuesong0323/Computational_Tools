```python
import pandas as pd
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/patrickcording/02807-comp-tools/master/docker/work/data/transactions.csv",encoding="utf-8")
```

```python
df
```

```python
# 这个是原来的代码，我们来一步一步分解它
city_country_mapping = df.groupby(['city'])['country'].apply(lambda grp: list(grp.value_counts().index)[0]).to_dict()
```

```python
# 从最基本的开始，这一行代码会count the number of occurrences of every city-country pair
# For every city, we only want to keep the country that has the highest number of occurrences.
# For example, for city Amadora, it has two corresponding countries: Portual which has occurred 2509 times; Portuga which
# has only occurred 3 times. We only want to keep the first country, i.e. Portual.
df.groupby(['city'])['country'].apply(lambda grp: grp.value_counts())
```

```python
# 在上一行的基础上加入index，可以发现country这一列的每一行，都变成了一个list。
# 有的list不止有一个国家。我们只想要每个list里的第一个国家，因为第一个国家是出现频率最高的国家。
df.groupby(['city'])['country'].apply(lambda grp: grp.value_counts().index)
```

```python
# 因此，在上一行的基础上加入[0]，取出每个list里出现频率最高的那个国家，当做是这个城市的国家名
df.groupby(['city'])['country'].apply(lambda grp: grp.value_counts().index[0])
```

```python
# 最后，把处理好的数据存进dictionary里
df.groupby(['city'])['country'].apply(lambda grp: grp.value_counts().index[0]).to_dict()
```
